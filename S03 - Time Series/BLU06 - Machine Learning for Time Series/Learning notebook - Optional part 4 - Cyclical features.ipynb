### Cyclical Features

When dealing with time series, one of the first things data scientists are tempted to do in terms of feature engineering is to extract features from time series timestamps. Why? For several reasons:
* In a lot of time series, like number of sales, number of patients or number of bookings for a hotel on each day there is a strong seasonality component. As such, information like what quarter, week or week day we are in is quite important.
* Unless you have the [time stone](https://bit.ly/2FND5RA) like [Dr. Strange](https://bit.ly/2Ije2uE) (or [Thanos](https://bit.ly/2I1lEP2), or [Adam Warlock](https://bit.ly/2jBftqy)) does, we can consider features extracted from the timestamp to be exogenous variables. So, not much thought is needed for these features...or is it?

Let's start with a dataset about [Monthly Av. residential electricity usage Iowa city 1971 â€“ 1979](https://bit.ly/2I0pU1k)

```python
data = utils.load_electricity_consumption_series()
data.plot();
```

```python
data.head(5)
```

as you can see, there is a strong seasonal component in the consumption as expected. As such, we can extract features from timestamps that relate to the seasonality component in order to allow an arbitrary regression model to capture the season effects:

* semester number
* trimester number
* month number
* quarter number

We can't go below month because, as you can see, the observations are made monthly and, as such, there is no observed data for each week or weekday. Let's extract those features.

```python
data_expanded = data.reset_index()
data_expanded.head(10)
```

```python
timestamps = data_expanded.month

data_expanded['month number'] = timestamps.dt.month
data_expanded['quarter number'] = timestamps.dt.quarter
data_expanded['trimester number'] = (timestamps.dt.month % 3) + 1
data_expanded['semester number'] = (timestamps.dt.month % 2) + 1

data_expanded.head(10)
```

I have a question for you: is January closer to December or to April? The obvious answer is "Of course it is closer, what kind of question is that?!". Well, the issue is that, for the machine, the mapping between month name and number is the following:

* January: 1
* April: 4
* December: 12

Meaning that the distance between January and April will be smaller than the distance between January and December. Given the fact that models like linear models, kNN, K-Means and neural networks are very sensitive to this difference, we need to preprocess these features we extracted. Usually, the first approach people take is to encode features like this using a one hot encoder (remember the dummy encoder? that's the name in scikit learn). But this type of encoding will remove precious information about the order of each unique value (i.e. January comes before February, March after February, ...).

Instead of using the features as they are or encoding them as if they were categorical, we can accept their **cyclical nature** and use encodings that leverage on that same nature. Imagine that we have observations on a hourly basis. When we think about hours, we usually think about clocks. And, as we all know, traditional clock design tends to be circular.

<img src="media/wall-clock.jpg" width="200">

in these kind of clocks, we **map** each daily **time measure** to a **point** on the **circle**. Can we do that sort of mapping using mathematical tools? 

Imagine that each daily time measure is a point on the circumference of a unit circle.

<img src="media/unit-circle.png">

For example, 00h00 would be at (1,0), 06h00 at (0,1), 12h00 at (-1,0) and 18h00 at (0,-1). If the time unit was an hour, our circle would be divided into 24 pieces. Now you are asking: what functions can I use to map the number of the hour ([0; 23]) to a point in the unit circle circumference, **in a continuous fashion**? For example, what would be the circumference point for 13h29m05s? Since we are using unit circles, we can apply **trigonometric functions** to hour time measures in order to map them into the circle.

<img src="media/unit-circle-with-sin-cos.gif">

You might be thinking "But...I don't have any angle and the picture you used has an angle...". True, we'll show you below how to get the angle.

All operations we will perform are measured in radians. Remember: 360 degrees, in radians, is equal to 2$\pi$.

First, let's create an example dataframe with a time range spanning 1 day, with hourly time stamps.

```python
example = pd.DataFrame(index=pd.date_range(start='February 1 2018', end='February 2 2018', freq='h'))
example.head(5)
```

Let's add the *hour* column

```python
example['hour'] = example.index.hour
example.head(5)
```

In order to turn the values of the *hour* column into an angle, we divide the full circle angle 2$\pi$ into the number of hours, that multiply by our hour value $k$.

$$\theta = \frac{2 k \pi}{N}$$

For example, for the mapping we created in the example dataframe, $k=0$ for 00h00, $k=3$ for 03h00. Now, you might be wondering "How do I map things like 00h30?". The answer is: well, it is half way between 00h00 and 01h00, which corresponds to $k=0$ and $k=1$, we should use $k=0.5$. Now, let's get the unit circle representation for the hour.

```python
example['hour (sin)'] = np.sin((2. * example['hour'] *  np.pi / 24))
example['hour (cos)'] = np.cos((2. * example['hour'] *  np.pi / 24))
```

Finally, let's look at the time series for these two new features.

```python
example['hour (sin)'].plot()
example['hour (cos)'].plot();
```
