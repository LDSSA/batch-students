### 2.2 Date formats 

Now you may be thinking _"hang on, was that just because the strings were exactly in the way pandas likes them?"_

It's a fair question, and well, I have to admit that it's partly true. Pandas' [`to_datetime`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) can infer formats, but it doesn't always do it. We'll see below how it works.

Let's put it to the test. We'll first define a little function to sanity check our dates.

```python
def sanity_check(dates):
    # go ahead Pandas, guess my date format! 
    inferred_dates = pd.to_datetime(dates)
    
    # Print out the results 
    print('Our first day is   5,    and was infered as %0.0f' % inferred_dates.iloc[0].day)
    print('Our first month is 4,    and was infered as %0.0f' % inferred_dates.iloc[0].month)
    print('Our first year is  2007, and was infered as %0.0f' % inferred_dates.iloc[0].year)
```

Let's start with an easy one.

```python
american_dates = pd.Series(['04/05/2007',  # <-- April 5th, 2007
                            '04/13/2006', 
                            '12/27/2014'])

sanity_check(american_dates)
```

Can we separate them with hyphens? 

```python
hyphen_separated_dates = pd.Series(['04-05-2007',  # <-- April 5th, 2007
                            '04-13-2006', 
                            '12-27-2014'])

sanity_check(hyphen_separated_dates)
```

Eh... English? 

```python
dates_in_english = pd.Series(['April 5th, 2007',  # <-- April 5th, 2007
                            'April 13th, 2006', 
                            'December 27th, 2014'])

sanity_check(dates_in_english)
```

Wow! So, European dates should be easy... right? 

```python
european_dates = pd.Series(['05/04/2007',   # <-- April 5th, 2007
                            '13/04/2006',
                            '27/12/2014'])
try:
    sanity_check(european_dates)
except ValueError as e:
    print(f"Value Error: {e}")
```

Wait... what? It was caught in our exception, because it was expecting the "%m/%d/%Y" format, but it saw 13 as the first value which triggered the exception.

It turns out pandas can infer lots of things, but Europe isn't its strength. Even though the second and third line clearly indicate that the month is in the middle (the 13'th can't be a month), it still gets confused. This is because the argument `dayfirst` is False by default and as the first date could be interpreted as dayfirst, this was is the option pandas chose.

What if we exchange the first two dates?

```python
european_dates_take_2 = pd.Series(['13/04/2006',
                            '05/04/2007',
                            '27/12/2014'])
try:
    sanity_check(european_dates_take_2)
except ValueError as e:
    print(f"Value Error: {e}")
```

Pandas again used the first date to infer the format, so this time it worked, but we got a warning and a very good suggestion.

Here is where line 2 of [The Zen of Python](https://www.python.org/dev/peps/pep-0020/#id3) comes in:
> Explicit is better than implicit

Let's heed the advice.

```python
inferred_dates = pd.to_datetime(european_dates, 
                                dayfirst=True)  # <--- explicit! 
```

```python
print('Our first day is   5,    and was infered as %0.0f' % inferred_dates.iloc[0].day)
print('Our first month is 4,    and was infered as %0.0f' % inferred_dates.iloc[0].month)
print('Our first year is  2007, and was infered as %0.0f' % inferred_dates.iloc[0].year)
```

By being explicit, we can parse arbitrarily crazy dates:

```python
dates_in_quackland = pd.Series(['05_quack_2007$04',   # <-- April 5th, 2007, in quack_timesystem
                                '13_quack_2006$04',    
                                '27_quack_2014$12'])

inferred_dates = pd.to_datetime(dates_in_quackland, 
                                format='%d_quack_%Y$%m')  # %d is day, %m is month,
                                                          # %Y is 4 digit year

print('Our first day is   5,    and was infered as %0.0f' % inferred_dates.iloc[0].day)
print('Our first month is 4,    and was infered as %0.0f' % inferred_dates.iloc[0].month)
print('Our first year is  2007, and was infered as %0.0f' % inferred_dates.iloc[0].year)
```

Believe me, when parsing different types of dates, this [page](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior) in the Python documentation with the accepted format codes will come in handy.

Geeks among us will be thinking _"That's all good and fine, but [real programmers](https://xkcd.com/378/) use time since epoch!"_

Well fear not, pandas has got you covered. 

```python
dev_time = inferred_dates.astype('int64')  # our inferred dates were datetime objects, remember? 
dev_time
```

And to convert back? 

```python
pd.to_datetime(dev_time)
```

We think in terms of people "per minute" or "per hour". 

### 2.4 Aggregation of time series data

So... Let's get our volume per minute! For this, we can use [resample](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html).

In our case, we'll resample to every 5 minutes, and take the sum (because we want to sum the volume in those 5 minutes). 

```python
data.loc[interval]['Volume_(Currency)'].resample('5 min').sum().plot()
plt.ylabel('Volume')
plt.title('Volume on 01/17/2018 between 1:30 and 4:30PM');
```

Wow! How much money (in dollars) was traded in the largest 10 minutes peak?

```python
money = data.loc[interval]['Volume_(Currency)'].resample('10 min').sum().max()

print('In 10 minutes, %0.1f dollars were traded in Bitcoin' % money)
```

We just took the sum, but if we were looking at prices, would that make sense? Probably not, we would resample and take the mean:

```python
data.resample('W').Close.mean().plot();  # the mean weekly closing prices 
                                         # 'W' is a so-called alias for week
plt.ylabel('Close price')
plt.title('Mean Close price since 2017');
```

Time... is... cool. 

What if we wanted to know the total amount of money that has been traded in this time span?

One way would simply be to sum it, but that doesn't give us any idea of how that total varied over time:

```python
data['Volume_(Currency)'].sum()
```

A cooler way to see this over time is to use the cumulative sum:

```python
data['Volume_(Currency)'].cumsum().plot();  # the total volume traded since the start 
plt.ylabel('Volume')
plt.title('Cumulative volume since 2017');
```

As you've learnt before, there are many cool [methods for groupby](https://pandas.pydata.org/pandas-docs/stable/reference/groupby.html). 

Let's say we want to know "what was the record for total volume per day, over time?" Naturally the record can only go up, and will have some "steps". 

```python
data['Volume_(Currency)'].resample('d').sum().cummax().plot(); 
plt.ylabel('Volume')
plt.title('Cumulative sum of volume since 2017')
```

Another useful method is the [rolling window](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rolling.html). It does what the name suggests: aggregate the previous X timepoints using a certain function, for instance the mean. It's very useful to smooth choppy timeseries into something less reactive to noise. 

We can choose to center the window (look back and forward), but in general, we only want to take into account information from the past, so we should use `center=False` (which is the default).

I think you get the picture. What's going on is that we're being extremely reactive to noise, and missing the underlying process. What is in fact going on is that we are in a free-fall, but it might not be obvious unless we look at the slightly broader picture. 

In other words, assuming there is an underlying process, we can assume the recent past should carry some weight. How much weight? A [rolling window](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.rolling.html) of weight! 

The first hour of Dec 18th 2017, as seen by traders:

```python
data.loc['Dec 18th 2017 00:00:00':'Dec 18th 2017 01:00:00', 'Weighted_Price'].plot();
plt.title('Weighted Price between 00:00 and 1:00 on 2017/12/18')
plt.ylabel('Weighted Price');
```

The first hour of Dec 18th 2017, as seen by a rolling window of 10 minutes:

```python
# This is just the raw data, so we can apply a rolling window on it.
first_hour = data.loc['Dec 18th 2017 00:00:00':'Dec 18th 2017 01:00:00', 'Weighted_Price']

# Notice the window size as a parameter of rolling, feel free to change that parameter.
# The center set to False. That's because we don't want to use data from the future! 
# Also notice how we use the mean. We can use many other functions. Try changing it! 
window_size = 10 
first_hour_rolling_window = first_hour.rolling(window=window_size, center=False).mean()
```

Now we'll plot the rolling mean and the original data together.

```python
first_hour_rolling_window.plot(color='b', label='rolling_window = %0.0f' % window_size);
first_hour.plot(label='raw data', alpha=.7, ls='-', color='orange');
plt.ylabel('Weigted Price')
plt.title('Weighted Price between 00:00 and 01:00 on 2017/12/18')
plt.legend();
```

As you can see, the rolling mean appears as a smoother version of the original signal, this can be a great help when looking at noisy signals where you just want to get an idea of the overall trend. Also note that when doing the rolling mean you get a delayed version of the original signal, due to the nature of how it is calculated.

( _Note: In case you are curious about what would have happened if you had interpreted the yellow line as a potential recovery... It doesn't end well._ )

Both rolling windows and resampling methods aggregate timeseries. If you still have doubts about the difference between them consult the section [Time-aware rolling vs. resampling](https://pandas.pydata.org/pandas-docs/version/1.1.5/user_guide/computation.html#time-aware-rolling-vs-resampling) of pandas user guide.

## 3. Missing data in time series

In the previous specialization, you've heard about the problem of missing data, and you guessed it right, for time series this may be an even more important problem. Given that the consecutive samples are dependent on each other, having missing timepoints may have a snowball effect. For example, you may be dependent on the current timestamp to predict the next one. If you're missing today's value, it will be hard to predict tomorrow's.

Let's see if we have any missing data in our dataset:

```python
data.isnull().sum()
```

It seems like we're missing some values on the Low column! Let's take a closer look:

```python
data.loc[data.isnull().Low]
```

They are all on the 11th of March. Let's also look at the plot.

```python
data.loc['2018-03'].Low.plot()
plt.ylabel('Low Price')
plt.title('Low price on March 2018');
```

There's the gap.

The way to solve this will be highly dependent on your particular problem. The simplest solution is to use the forward fill, basically fill the missing values using the last existing value. This method makes strong assumptions that won't probably make sense to most datasets, and you may indeed be adding noise to your data. 

But for example, there are industrial signals that work on-change. This means that they only register values when they detect changes and all the rest are NaNs. In this cases using forward filling is the appropriate approach.

```python
interval = slice('2018-03-10','2018-03-11')
```

```python
data.loc[interval].Low.ffill().plot()
plt.ylabel('Low price')
plt.title('Low price between 2018-03-10 and 2018-03-11')
```

Not great. For this particular case, a simple linear interpolation could be slightly better.

```python
data.loc[interval].Low.interpolate().plot()
plt.ylabel('Low price')
plt.title('Low price between 2018-03-10 and 2018-03-11');
```

But we have a multivariate time series. We should be able to take advantage of the relationship between the variables to estimate the missing data. For that we can use sklearn's `KNNImputer` which is usually a simple yet elegant way to solve the problem when we don't really have domain knowledge to do anything better. It just may take a while to run.

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer() # To obtain better results we may want to have more neighbors, in a real setting
                       # this is a parameter that should be tunned.
    
data_imputed = imputer.fit_transform(data)
df_imputed = pd.DataFrame(data_imputed, index=data.index, columns=data.columns)
```

```python
df_imputed.loc[interval].Low.plot()
plt.ylabel('Low Price')
plt.title('Low price between 2018-03-10 and 2018-03-11');
```

Well, at least it looks a bit more like the rest of the data! Not surprising if we look at the variation of the open, high, and close price.

```python
data.loc[interval][['Open','Low','High','Close']].plot();
```

Again, the idea is to introduce you to methods you may want to use in the future, but as in all cases, missing data imputation is highly specific. Do remember that when using KNN imputation in a real scenario you should fit the transformer on a training set and then do the transformations in the validation and test sets.

    * Remember pandas is really helpful here as you can even do something like `dataframe.loc['Jan 2018']` or even ranges `dataframe.loc['01/15/2018':'01/22/2018']`
    * If you're going to analyse the same interval several times, then using `slice()` can help a lot.
*  How to resample, e.g: `.resample('5min').sum()`. Don't forget that after resampling we need an aggregation function! 
    * new aggregation functions that we've talked about: 
        * `cumsum()`
        * `cummax()`
* We have also learnt how to calculate the difference between timepoints, using `diff()`
* Rolling windows: `.rolling()` that allows us to specify the rolling window size and the aggregation function.
* Missing value imputation using forward filling or KNN imputation.
