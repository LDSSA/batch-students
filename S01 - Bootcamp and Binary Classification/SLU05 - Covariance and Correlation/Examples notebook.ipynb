# SLU 05 - Covariance and Correlation: Example notebook

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
from matplotlib import pyplot as plt 

# California housing dataset
from sklearn.datasets import fetch_california_housing
```

In this notebook, you will see examples on how to calculate:

- Covariance 
- Pearson correlation
- Spearman correlation
- Correlation matrix

You will also find examples of:

- Spurious correlations
- Observational and Experimental Data

## Loading and previewing the data

```python
d_X, d_Y = fetch_california_housing(return_X_y=True, as_frame=True)
housing_data = d_X.join(d_Y)
housing_data.head()
```

## Covariance

Calculating the covariance between `AveRooms` and `AveBedrms`:

```python
housing_data['AveRooms'].cov(housing_data['AveBedrms'])
```

## Pearson correlation

Calculating the Pearson correlation between `AveRooms` and `AveBedrms`:

```python
housing_data['AveRooms'].corr(housing_data['AveBedrms'], method='pearson')
```

## Spearman correlation

Calculating the Spearman correlation between `AveRooms` and `AveBedrms`:

```python
housing_data['AveRooms'].corr(housing_data['AveBedrms'], method='spearman')
```

## Correlation matrix

Using the standard Pandas method:

```python
housing_data.corr()
```

With visualization:

```python
sns.heatmap(housing_data.corr(), annot=True);
```
